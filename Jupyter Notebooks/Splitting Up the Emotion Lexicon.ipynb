```python
import pickle
import pandas as pd
import numpy as np
import re
```

```python
with open('emotionScores.pickle', 'rb') as f:
    lexicon = pickle.load(f)
```

```python
lexicon.head()
```

```python
lexicon['Part_of_Speech'].unique()
```

```python
anger_words = []
fear_words = []
joy_words = []
sadness_words = []
surprise_words = []

for i in range(len(lexicon)):
    if lexicon['Part_of_Speech'][i] != 'r':
        output = list(lexicon[['Afraid', 'Amused', 'Angry', 'Annoyed', 'Dont_Care', 'Happy', 'Inspired', 'Sad']].loc[i].values)
        result = output.index(max(output))
        new = sorted(output)
        while result == 1 or result == 3 or result == 4:
            new = new[:-1]
            result = output.index(max(new))
        if result == 0:
            fear_words.append(lexicon['Armenian'][i])
        elif result == 2:
            anger_words.append(lexicon['Armenian'][i])
        elif result == 5:
            joy_words.append(lexicon['Armenian'][i])
        elif result == 6:
            surprise_words.append(lexicon['Armenian'][i])
        elif result == 7:
            sadness_words.append(lexicon['Armenian'][i])
```

```python
string = ' '.join(listed_text)
```

```python
anger_counts = []
fear_counts = []
joy_counts = []
sadness_counts = []
surprise_counts = []

for i in range(len(anger_words)):
    anger_counts.append(string.count(anger_words[i]))

for i in range(len(fear_words)):
    fear_counts.append(string.count(fear_words[i]))
    
for i in range(len(joy_words)):
    joy_counts.append(string.count(joy_words[i]))

for i in range(len(sadness_words)):
    sadness_counts.append(string.count(sadness_words[i]))

for i in range(len(surprise_words)):
    surprise_counts.append(string.count(surprise_words[i]))
```

```python
anger_contributions = []
fear_contributions = []
joy_contributions = []
sadness_contributions = []
surprise_contributions = []

for i in range(len(anger_counts)):
    x = anger_counts[i]
    anger_contributions.append(1 / (1 + np.exp(-x)))

for i in range(len(fear_counts)):
    x = fear_counts[i]
    fear_contributions.append(1 / (1 + np.exp(-x)))
    
for i in range(len(joy_counts)):
    x = joy_counts[i]
    joy_contributions.append(1 / (1 + np.exp(-x)))

for i in range(len(sadness_counts)):
    x = sadness_counts[i]
    sadness_contributions.append(1 / (1 + np.exp(-x)))

for i in range(len(surprise_counts)):
    x = surprise_counts[i]
    surprise_contributions.append(1 / (1 + np.exp(-x)))
```

```python
ang = {'Word':anger_words, 'Count':anger_counts, 'Contribution':anger_contributions}
fea = {'Word':fear_words, 'Count':fear_counts, 'Contribution':fear_contributions}
jo = {'Word':joy_words, 'Count':joy_counts, 'Contribution':joy_contributions}
sad = {'Word':sadness_words, 'Count':sadness_counts, 'Contribution':sadness_contributions}
sur = {'Word':surprise_words, 'Count':surprise_counts, 'Contribution':surprise_contributions}

anger = pd.DataFrame(ang)
fear = pd.DataFrame(fea)
joy = pd.DataFrame(jo)
sadness = pd.DataFrame(sad)
surprise = pd.DataFrame(sur)
```
