```python
import pickle
import pandas as pd
import numpy as np
import re
```

```python
with open('sentimentScores_final.pickle', 'rb') as f:
    lexicon = pickle.load(f)
```

```python
lexicon.head()
```

```python
lexicon['Part_of_Speech'].unique()
```

```python
positive_words = []
negative_words = []
positive_adverb = []
negative_adverb = []

for i in range(len(lexicon)):
    if lexicon['Part_of_Speech'][i] != 'r':
        if lexicon['Positive'][i] > lexicon['Negative'][i]:
            positive_words.append(lexicon['Armenian'][i])
        elif lexicon['Positive'][i] < lexicon['Negative'][i]:
            negative_words.append(lexicon['Armenian'][i])
    else:
        if lexicon['Positive'][i] > lexicon['Negative'][i]:
            positive_adverb.append(lexicon['Armenian'][i])
        elif lexicon['Positive'][i] < lexicon['Negative'][i]:
            negative_adverb.append(lexicon['Armenian'][i])
```

```python
positive_counts = []
negative_counts = []

for i in range(len(positive_words)):
    positive_counts.append(string.count(positive_words[i]))

for i in range(len(negative_words)):
    negative_counts.append(string.count(negative_words[i]))
```

```python
positive_counts
```

```python
negative_counts
```

```python
positive_contributions = []
negative_contributions = []

for i in range(len(positive_counts)):
    x = positive_counts[i]
    positive_contributions.append(1 / (1 + np.exp(-x)))

for i in range(len(negative_counts)):
    x = negative_counts[i]
    negative_contributions.append(1 / (1 + np.exp(-x)))
```

```python
positive_contributions
```

```python
negative_contributions
```

```python
pos = {'Word':positive_words, 'Count':positive_counts, 'Contribution':positive_contributions}
neg = {'Word':negative_words, 'Count':negative_counts, 'Contribution':negative_contributions}

positive = pd.DataFrame(pos)
negative = pd.DataFrame(neg)
```

```python
with open('positive_contribution.pickle', 'wb') as f:
    pickle.dump(positive, f)

with open('negative_contribution.pickle', 'wb') as f:
    pickle.dump(negative, f)
```

```python
with open('positive_adverb.pickle', 'wb') as f:
    pickle.dump(positive_adverb, f)
with open('negative_adverb.pickle', 'wb') as f:
    pickle.dump(negative_adverb, f)
```
